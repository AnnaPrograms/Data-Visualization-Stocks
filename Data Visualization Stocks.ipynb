```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("data/AAPL.csv", parse_dates = ["Date"])

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize = (13, 7))

# Customizes canvas
fig.tight_layout(pad=5)
fig.suptitle('Stock Prices of Apple inc.')

linethic = 1

# Plot 1
ax1.plot(df.Date, df["Close"], color="crimson", linewidth=linethic, linestyle="-", label = "Close")
ax1.set_title("Close Price Trend")
ax1.set_ylabel("Close")
ax1.set_xlabel("Date")
ax1.legend(title="Price")

# Plot 2
ax2.plot(df.Date, df["Open"], color="blue", linewidth=linethic, linestyle="-", label = "Open")
ax2.set_title("Open Price Trend")
ax2.set_ylabel("Close")
ax2.set_xlabel("Date")
ax2.legend(title="Price")


# Plot 3
ax3.plot(df.Date, df["High"], color="green", linewidth=linethic, linestyle="-", label = "High")
ax3.set_title("High Price Trend")
ax3.set_ylabel("Close")
ax3.set_xlabel("Date")
ax3.legend(title="Price")

# Plot 4
ax4.plot(df.Date, df["Low"], color="olive", linewidth=linethic, linestyle="-", label = "Low")
ax4.set_title("Low Price Trend")
ax4.set_ylabel("Close")
ax4.set_xlabel("Date")
ax4.legend(title="Price")

```
